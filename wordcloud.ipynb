```python
import json

from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
with open(r"E:\Datasets\Acoustic_model\vocab.json", "r") as f:
    vocab = json.load(f)

print(vocab)
```

```python
t = "~".join(list(vocab.keys())[5:])
t
```

```python
plt.figure(figsize=(10,10))
wordcloud = WordCloud(width=1500, height=500, background_color="white").generate(t)
plt.imshow(wordcloud)
plt.axis("off")
plt.savefig("vocab_cloud.png")
plt.show()
```
